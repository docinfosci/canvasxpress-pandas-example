```python
# Import CanvasXpress and its global graph function.
from canvasxpress.canvas import CanvasXpress
from canvasxpress.plot import graph

# Import pandas, which provides DataFrame.
import pandas

# Import requests for HTTP access and io for in-memory file-like storage.
import requests
import io
```

```python
# Import the primary data marix.
data_url = "https://www.canvasxpress.org/data/r/cX-generic-dat.txt"
data_raw = requests.get(data_url).content
data_df = pandas.read_csv(
    io.StringIO(data_raw.decode('utf-8')),
    sep="\t",
    index_col=0,
)

# Inspect the data.
data_df.head()
```

```python
# Import smps (column) annotations / metadata.
sample_annot_url = "https://www.canvasxpress.org/data/r/cX-generic-smp.txt"
sample_annot_raw = requests.get(sample_annot_url).content
sample_annot_df = pandas.read_csv(
    io.StringIO(sample_annot_raw.decode('utf-8')),
    sep="\t",
    index_col=0,
)

# Inspect the data.
sample_annot_df.head()
```

```python
# Import vars (row) annotations / metadata.
variable_annot_url = "https://www.canvasxpress.org/data/r/cX-generic-var.txt"
variable_annot_raw = requests.get(variable_annot_url).content
variable_annot_df = pandas.read_csv(
    io.StringIO(variable_annot_raw.decode('utf-8')),
    sep="\t",
    index_col=0,
)

# Inspect the data.
variable_annot_df.head()
```
